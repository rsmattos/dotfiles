```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
```

```python
mldyn_traj = np.empty(351)
ml_time = np.empty(351)

with open('run_traj/TRAJ_fast/positions.dat') as f:
    lines = f.readlines()

    for idx, line in enumerate(lines):
        word = line.split()

        if word[0] == '#':
            continue

        ml_time[idx-1] = word[0]
        mldyn_traj[idx-1] = word[1]

ml_time = ml_time / 41.34137221718
mldyn_traj = mldyn_traj * 0.52917720859
```

```python
h5 = h5py.File('run_traj/TRAJ_fast/dyn.h5')

nx_traj = h5['particles/all/position/value'][:,0,0]
nx_time = h5['particles/all/position/time'][:]
```

```python
plt.plot(ml_time, mldyn_traj, label='mldyn')
plt.plot(nx_time, nx_traj, label='nx')
plt.legend()
plt.title('fast traj mom 25 au')
```
